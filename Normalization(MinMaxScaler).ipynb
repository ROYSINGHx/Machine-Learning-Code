```python
import pandas as pd
import numpy as np

df = pd.read_csv('/content/wine_data.csv', header=None , usecols = [0,1,2])
df.columns = ['Class label', 'Alcohol', 'Malic acid']
```

```python
df.head()
```

```python
x = df.drop('Class label',axis=1)
y = df['Class label']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.3 , random_state=2)
```

```python
from sklearn.preprocessing import MinMaxScaler
mns = MinMaxScaler()
mns.fit(x_train)
```

```python
x_train_ss = mns.transform(x_train)
x_test_ss = mns.transform(x_test)
```

```python
x_train_ss
```

```python
x_train_sss = pd.DataFrame(x_train_ss , columns=x_train.columns)
x_test_sss = pd.DataFrame(x_test_ss , columns=x_test.columns)
```
