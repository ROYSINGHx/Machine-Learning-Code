```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('/content/Social_Network_Ads.csv')
```

```python
df=df.iloc[:,2:]
```

```python
df.head()
```

```python
x = df.drop('Purchased',axis=1)
y = df['Purchased']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.3 , random_state=2)
```

```python
from sklearn.preprocessing import StandardScaler
sl = StandardScaler()
```

```python
sl.fit(x_train)#mean and variance
x_train_ss = sl.transform(x_train)#fit into in formula
x_test_ss =sl.transform(x_test)
```

```python
x_train_ss
```

```python
x_train_sss = pd.DataFrame(x_train_ss , columns=x_train.columns)
x_test_sss = pd.DataFrame(x_test_ss , columns=x_test.columns)
```
